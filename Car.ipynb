# Predication

```python
X = df.drop(columns = ['price'],axis = 1)
y = df['price']
```

```python
X
```

# convert into numeric
# one-hot Encoding

```python
df.columns
```

```python
X_one_encoded = pd.get_dummies(X, columns = ['model','transmission','fuelType'],drop_first = True)
```

```python
X_one_encoded
```

```python
X_one_encoded = X_one_encoded.astype(int)
```

```python
X_one_encoded 
```

# labelEncoding

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
columns = ['model','transmission','fuelType']
Xlabel = X
for i in columns:
    Xlabel[i] = encode.fit_transform(Xlabel[i])
```

```python
Xlabel
```

```python
# standardScaler
from sklearn.preprocessing import StandardScaler
numeric_cols = ['year','mileage','tax','mpg','engineSize']
scaler = StandardScaler()
X_one_encoded[numeric_cols] = scaler.fit_transform(X_one_encoded[numeric_cols])
```

```python
X_one_encoded
```

```python
Xlabel.columns
```

```python
Xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(Xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
Xlabel
```

# Create Model By Using One-Hot Encoding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error ,mean_squared_error , r2_score
```

```python
# LinearRegression
X_train, X_test, y_train, y_test = train_test_split( X_one_encoded, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Model Accurate
r2 = r2_score(y_test,y_pred)
r2
```

```python
# Model Adjusted
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 -((1-r2)*(n-1))/(n-p-1)
print("Adjusted R2 Score:",adjusted_r2)
```
